```python
import matplotlib.pyplot as plt
import networkx as nx

# Création du graphe
G = nx.DiGraph()

# Ajout des nœuds
G.add_nodes_from(['A', 'B', 'C', 'D'])

# Ajout des arêtes
G.add_edge('A', 'B')
G.add_edge('B', 'C')
G.add_edge('B', 'D')
G.add_edge('D', 'C')
G.add_edge('C', 'A')
G.add_edge('B', 'B')
G.add_edge('D', 'A')

# Positionnement des nœuds
pos = {'A': (1, 1), 'B': (2, 2), 'C': (3, 2), 'D' : (3,3)}

# Tracé du graphe
plt.figure(figsize=(5, 4))
nx.draw(G, pos, with_labels=True, node_size=800, node_color='skyblue', font_size=15, font_color='black', font_weight='bold')
plt.title("Graphe dirigé avec 4 nœuds et des liens")
plt.show()
```
